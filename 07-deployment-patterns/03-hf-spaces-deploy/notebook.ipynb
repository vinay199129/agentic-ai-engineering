# HF Spaces deploy

Three artefacts make a deployable Space:

1. `space/README.md` with HF front-matter that declares the SDK + app file.
2. `space/app.py` Streamlit entrypoint.
3. `space/requirements.txt` pinned for the Space runtime.

GitHub Actions (`workflow.yml`) takes care of every redeploy.

````python
import pathlib, re, yaml
txt = pathlib.Path('space/README.md').read_text()
fm = yaml.safe_load(re.match(r'^---\n(.*?)\n---', txt, re.DOTALL).group(1))
fm
````

## CI deploy in one command

```bash
huggingface-cli upload $HF_USERNAME/$HF_SPACE_NAME ./space . --repo-type=space
```

Driven by `workflow.yml`; secrets live in repo settings (never in the code).
