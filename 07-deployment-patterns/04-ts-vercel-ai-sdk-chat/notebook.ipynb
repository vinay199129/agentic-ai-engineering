Two-tier deploy: a **stateful Python agent** (FastAPI, leaf 00) and a **stateless edge UI** (Next.js on Vercel). The edge route proxies SSE so the browser sees a same-origin stream — no CORS, no cookies on the cross-origin path.

## Why hand-roll the SSE reader?

`useChat` from `ai/react` expects OpenAI-shaped completion frames. Our agent backend emits **graph events** (`node_complete`, `interrupt`, …) — a different schema. A 30-line reader keeps the contract honest and lets us surface the interrupt as a UI affordance.

## Deploy

* Push to a repo connected to Vercel.
* Set `AGENT_BACKEND_URL` (server) and `NEXT_PUBLIC_AGENT_BACKEND_URL` (browser) to point to your prod FastAPI service.
* The API route runs on the Edge runtime — global low-latency proxying for free.
