# Streamlit demo template

Generic chat UI + tool-trace expander. Demos plug in by passing a `solve_fn` into `run(...)`. Default solver is the in-repo MCP agent so the page is functional out of the box without API keys.

Launch with:

```powershell
uv run streamlit run 07-deployment-patterns/01-streamlit-demo-template/app.py
```

````python
import os, sys, pathlib
ROOT = pathlib.Path.cwd()
while not (ROOT / 'shared').exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))
sys.path.insert(0, str(ROOT / '07-deployment-patterns' / '01-streamlit-demo-template'))
sys.path.insert(0, str(ROOT / '06-mcp'))
os.environ.setdefault('LLM_CACHE_ONLY', '1')
import app
result = app.solve_demo('What does RA-MoE do?')
print('answer  :', result['answer'][:200])
print('# steps :', len(result['trace']))
````

## Customising for your own demo

Create a sibling script:

```python
from app import run
from my_agent import solve
run(solve_fn=solve, title='RAG over our docs')
```

and launch with `streamlit run my_demo.py`.
