# Docker Compose stack — local agent dev environment

Four services in one file: FastAPI agent, Postgres+pgvector, Redis, Langfuse. Bring it up with `docker compose up -d`; tear it down with `docker compose down -v`.

```python
import pathlib, yaml
compose = yaml.safe_load(pathlib.Path('compose.yaml').read_text())
list(compose['services'])
```

```python
for name, svc in compose['services'].items():
    hc = '✅' if 'healthcheck' in svc else '⚠️'
    print(f"{name:10s} {svc.get('image','-'):40s} {hc}")
```

## Production migration path

* Move Postgres to managed (Supabase, Neon, RDS) — keep `pgvector` extension.
* Move Redis to managed (Upstash, MemoryDB).
* Run Langfuse on Fly.io / Render, or use the hosted version.
* The `api` service becomes a container image you build in CI and deploy to ECS / Cloud Run / Fly Machines.
