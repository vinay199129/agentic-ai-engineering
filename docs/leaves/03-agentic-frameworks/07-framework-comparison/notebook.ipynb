# Framework comparison — same task, seven implementations

Each sibling leaf in `03-agentic-frameworks/` solves the same canonical task (research and summarise a paper with citations) over the same three tools. This notebook joins their `eval-snapshot.json` files with a hand-curated capability matrix to produce a side-by-side leaderboard.

```python
import os, sys, pathlib, json
ROOT = pathlib.Path.cwd()
while not (ROOT / 'shared').exists() and ROOT != ROOT.parent:
    ROOT = ROOT.parent
sys.path.insert(0, str(ROOT))
sys.path.insert(0, str(pathlib.Path.cwd()))
from eval import FEATURE_MATRIX, PICK_WHEN, _load_sibling_snapshots, _render_leaderboard
metrics = _load_sibling_snapshots()
for tech, ms in metrics.items():
    print(f'{tech:<32}  ' + '  '.join(f'{k}={v}' for k, v in ms.items()))
```

## The leaderboard

```python
print(_render_leaderboard(metrics, FEATURE_MATRIX))
```

## How to read this

* The **metric columns** are identical-shape because every leaf calls the shared `run_evaluation(...)` helper in `task.py`. The metrics measure framework *plumbing*, not LLM quality.
* The **capability matrix** is opinionated. Where a framework supports a feature *by convention* (e.g. CrewAI's role/goal becomes a kind of typed I/O) we mark `~`.
* The **pick-when** list maps task shape → framework. It's the question to ask before reading individual leaves: *what does my agent need to do?*
